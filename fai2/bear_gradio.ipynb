```python
#|export
import gradio as gr
from fastai.vision.widgets import *
from fastai.vision.all import *
```

```python
path = Path(".")
path.ls(file_exts='export.pkl')
path
```

```python
#|export
learn_inf = load_learner(Path('export.pkl'))
```

```python
#|export
categories = learn_inf.dls.vocab
def classify_image(img):
    pred,idx,prob = learn_inf.predict(img)
    return dict(zip(categories,map(float,prob)))
```

```python
#|export
image = gr.Image(width=224, height=224)
labels = gr.Label()
examples = ['t1.jpeg', 'd.jpeg']
demo = gr.Interface(fn=classify_image, inputs=image, outputs=labels, examples=examples)
demo.launch() 
```
