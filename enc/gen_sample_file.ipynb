```python
from Crypto.Cipher import AES
from Crypto import Random
```

```python
SAMPLES=10 * 1000
```

```python
AES_BITS=128
AES_BYTES=int(AES_BITS/8)
print(f"{AES_BYTES=}")
```

```python
ctxts=list()
ptxts=list()
keys=list()
```

```python
for i in range(0, SAMPLES):
    key=Random.get_random_bytes(AES_BYTES)
    ptxt=Random.get_random_bytes(AES_BYTES)
    if ptxt in ptxts:
        continue
    cipher = AES.new(key, AES.MODE_ECB)
    ctxt = cipher.encrypt(ptxt)
    ptxts.append(ptxt)
    ctxts.append(ctxt)
    keys.append(key)
```

```python
list(map(lambda i : print(ptxts[i].hex(), ctxts[i].hex(), keys[i].hex()), range(0, SAMPLES)))
```

```python
with open("sample.txt", "w") as sf :
    list(map(lambda i : sf.write(ptxts[i].hex() + "," + ctxts[i].hex() + "," + keys[i].hex() + "\n"), range(0, SAMPLES)))
```
